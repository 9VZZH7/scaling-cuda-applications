# Outlook

Congratulations on finishing the course!

If there is still time left, you might want to challenge yourself.
For example:
* Implement efficient and hierarchical device reductions to track the aggregate temperature in each time step.
  * Perform an efficient reduction of the temperature as part of the main `stencil2D` kernel.
  * Use CUB or implement your own hierarchical reduction - warp or block level, then device level.
  * Add another layer of hierarchy - reduce across devices using one of the following ideas (or something else entirely):
    * MPI or NVSHMEM reductions in device memory.
    * NCCL reductions.
    * Atomics *across* devices with P2P access or NVSHMEM.

* Implement a 2D domain decomposition.
  * Adapt data structures, index sets, and copy operations.
  * Update the file output.
  * Buffer communication data in the x-direction in communication buffers.
  * Fuse writing to and reading from communication buffers with the main `stencil2D` kernel.

* Implement parallel file output with
  * MPI I/O.
  * a library of your choice.
